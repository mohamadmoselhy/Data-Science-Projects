```python
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
```

Euclidean Distance funcation

```python
# Defining Euclidean Distance funcation:
# Measures the shortest (direct) distance between two points (like a straight line).
# Uses the Pythagorean theorem to compute the distance.

def euclidean_distance(point1,point2):
    return np.sqrt(np.sum ((np.array(point2) - np.array(point1))**2))
```

Manhattan Distance

```python
# Defining Manhattan Distance funcation:
# Measures distance along axes (like moving in a grid).
# Does not consider diagonal movement, only horizontal & vertical steps.

def manhattan_distance(point1,point2):
    return np.sum(np.abs(np.array(point2) - np.array(point1) ))
```

```python
# Example:
print(euclidean_distance((1,2),(3,6)))
print(manhattan_distance((1,2),(3,6)))
```

```python
def minkowski_distance(point1, point2, p):
    return np.sum(np.abs(np.array(point1) - np.array(point2)) ** p) ** (1 / p)
```

```python
# Example usage
print("Euclidean Distance:", euclidean_distance((1, 2), (3, 6)))
print("Manhattan Distance:", manhattan_distance((1, 2), (3, 6)))
print("Minkowski Distance (p=1):", minkowski_distance((1, 2), (3, 6), 1))
print("Minkowski Distance (p=2):", minkowski_distance((1, 2), (3, 6), 2))
```

✅ Use Euclidean Distance for most KNN problems.
✅ Consider Manhattan Distance if data is grid-like or high-dimensional.
✅ For flexibility, Minkowski Distance allows tuning between both.

KNN Prediction Function:

```python
def knn_predict(TraningData,TraningLable,TestPoint,k):
    Distance = []
    for i in range(len(TraningData)):
        DistanceBetwwenPoint = euclidean_distance(TestPoint,TraningData[i])
        Distance.append((DistanceBetwwenPoint , TraningLable[i]))
    
    Distance.sort( key=lambda x: x[0])
    return Counter([label for _, label in Distance[:k]]).most_common(1)[0][0]
```

```python
# Training data and labels
training_data = [[1, 2], [2, 3], [3, 4], [6, 7], [7, 8], 
                 [8, 9], [10, 12], [12, 14], [14, 15], [15, 17]]
training_labels = ['A', 'A', 'B', 'A', 'A', 'A', 'B', 'B', 'A', 'B']

# Separate points by class
group_A = [p for p, l in zip(training_data, training_labels) if l == 'A']
group_B = [p for p, l in zip(training_data, training_labels) if l == 'B']

# Extract x, y coordinates
x_A, y_A = zip(*group_A)
x_B, y_B = zip(*group_B)

# Test point
test_point = [8, 9]

# Plot data
plt.scatter(x_A, y_A, color='blue', label="Class A", marker='o')
plt.scatter(x_B, y_B, color='red', label="Class B", marker='x')
plt.scatter(*test_point, color='green', label="Test Point", marker='D', s=100)  # Test point

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Training Data & Test Point")
plt.legend()
plt.grid(True)
plt.show()
k = 3
```

```python
print(knn_predict(training_data,training_labels,test_point,k))
```
